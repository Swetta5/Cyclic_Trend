```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Starbucks Sales Data (Monthly Sales Over 2 Years)
data = {
    'Date': pd.date_range(start='2022-01', periods=24, freq='ME'),  # Changed 'M' to 'ME'
    'Sales': [320, 350, 400, 370, 390, 450, 470, 430, 410, 460, 490, 500,  # Year 1
              520, 540, 600, 580, 590, 650, 670, 640, 610, 660, 690, 700]   # Year 2
}

# Create DataFrame
df = pd.DataFrame(data)
df.set_index('Date', inplace=True)  # Set Date as index for time-series analysis

# Calculate 3-month Moving Average (for smoothing cyclic trend)
df['3-Month MA'] = df['Sales'].rolling(window=3).mean()

# Plot Sales Data and Cyclic Trend
plt.figure(figsize=(8,5))
plt.plot(df['Sales'], marker='o', label='Actual Sales')
plt.plot(df['3-Month MA'], linestyle='dashed', color='red', label='Cyclic Trend (3-Month MA)')

plt.title("Starbucks Cyclic Trend Analysis")
plt.xlabel("Time (Months)")
plt.ylabel("Monthly Sales")
plt.legend()
plt.grid()
plt.show()
```
